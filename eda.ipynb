```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import Data

```python
data = pd.read_csv('1/IPIP-FFM-data-8Nov2018/data-final.csv')
```

Clean into row format

```python
rows = []
cols = data.columns[0].split('\t')
for index in range(len(data)):
    value = data.iloc[index, 0]
    row = (value.split('\t'))
    rows.append(row)
processed_data = pd.DataFrame(rows, columns=cols)
```

Filter for only 1 user per entry (more accurate data)

```python
# Filter the data where 'ipc' column is equal to 1
processed_data = processed_data[processed_data['IPC'] == '1']
```

```python
country_counts = processed_data["country"].value_counts()
print(country_counts)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct="%1.1f%%", startangle=140, colors=plt.cm.Paired.colors)
plt.title("Distribution of Countries")
plt.show()
```
